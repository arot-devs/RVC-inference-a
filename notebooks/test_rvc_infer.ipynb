```python
import os
os.environ['RVC_MODELDIR']='/root/dev/rvc-webui/assets/weights' #where model.pth files are stored.
os.environ['RVC_INDEXDIR']='/root/dev/rvc-webui/assets/indices' #where model.index files are stored.

#the audio output frequency, default is 44100.
# os.environ['RVC_OUTPUTFREQ']='44100'
os.environ['RVC_OUTPUTFREQ']='48000'

#If the output audio tensor should block until fully loaded, this can be ignored. But if you want to run in a larger torch pipeline, setting to False will improve performance a little.
os.environ['RVC_RETURNBLOCKING']='True'
```

```python
# weight: https://huggingface.co/trojblue/rvc-kanade-voice/blob/main/_weights_unsorted/knd48k-mp3_e720_s85680.pth
# indice: https://huggingface.co/trojblue/rvc-kanade-voice/blob/main/kanade_mp3_all_v2_48k.index

from inferrvc import RVC
rvc_class = RVC('knd48k-mp3_e720_s85680.pth', index='kanade_mp3_all_v2_48k')

print()
print(rvc_class.name)
print('Paths',rvc_class.model_path,rvc_class.index_path)
```

```python
from inferrvc import load_torchaudio


SAMPLE_AUDIO = '/root/dev/datasets/sample_audios/audio (9)_kkr_vits.wav'

aud,sr = load_torchaudio(SAMPLE_AUDIO)

paudio1=rvc_class(aud,f0_up_key=1,output_device='cpu',output_volume=RVC.MATCH_ORIGINAL,index_rate=.95)
# paudio2=obama(aud,5,output_device='cpu',output_volume=RVC.MATCH_ORIGINAL,index_rate=.9)

import soundfile as sf

sf.write(f"{SAMPLE_AUDIO}_convert.wav",paudio1,48000)
# sf.write('path/to/audio_obama.wav',paudio2,44100)
```
